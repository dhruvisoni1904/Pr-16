```python
import pandas as pd
import json

# Load JSON file
with open("job_salary_converted.json", "r") as f:
    data = json.load(f)

print("Step 2 Output (Sample JSON):")
print(data[:2])
```

```python
df_structured = pd.json_normalize(data)

print("Step 3 Output (Flattened):")
print(df_structured.head())
```

```python
df_final = df_structured.copy()

print("\nStep 4 Output (Fully Structured):")
print(df_final.head())
```

```python
print("\n--- Statistics ---")
print(df_final.describe())
```

```python
df_final.to_csv("structured_output.csv", index=False)

print("\nFile 'structured_output.csv' is ready for Excel.")
```
